```python
# Select the regression models to evaluate
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from xgboost import XGBRegressor
```

```python
# Preprocess the data
X_train = train_data.drop(['price'], axis=1)
y_train = train_data['price']

X_test = test_data.drop(['price'], axis=1)
y_test = test_data['price']
```

```python
X_train = train_data.drop(['price'], axis=1)
y_train = train_data['price']
```

```python
X_test = test_data.drop(['price'], axis=1)
y_test = test_data['price']
```

```python
# Fit the regression models to the training data
xgb_model = XGBRegressor()
xgb_model.fit(X_train, y_train)
```

```python
linear_model = LinearRegression()
linear_model.fit(X_train, y_train)
```

```python
rf_model = RandomForestRegressor()
rf_model.fit(X_train, y_train)
```

```python
svr_model = SVR()
svr_model.fit(X_train, y_train)
```

```python
# Evaluate the performance of each model
from sklearn.metrics import r2_score, mean_squared_error
```

```python
y_pred_linear = linear_model.predict(X_test)
linear_r2 = r2_score(y_test, y_pred_linear)
linear_mse = mean_squared_error(y_test, y_pred_linear)
```

```python
y_pred_rf = rf_model.predict(X_test)
rf_r2 = r2_score(y_test, y_pred_rf)
rf_mse = mean_squared_error(y_test, y_pred_rf)
```

```python
y_pred_svr = svr_model.predict(X_test)
svr_r2 = r2_score(y_test, y_pred_svr)
svr_mse = mean_squared_error(y_test, y_pred_svr)
```

```python
y_pred_xgb = xgb_model.predict(X_test)
xgb_r2 = r2_score(y_test, y_pred_xgb)
xgb_mse = mean_squared_error(y_test, y_pred_xgb)
```

```python
print("Linear Regression R-squared:", linear_r2)
print("Random Forest R-squared:", rf_r2)
print("Support Vector Regression R-squared:", svr_r2)
print("XGBoost R-squared:", xgb_r2)
```

```python
print("Linear Regression MSE:", linear_mse)
print("Random Forest MSE:", rf_mse)
print("Support Vector Regression MSE:", svr_mse)
print("XGBoost MSE:", xgb_mse)
```

```python
# Create a table of feature importance for each model
linear_importance = pd.DataFrame({
    'Feature': X_train.columns,
    'Coefficient': linear_model.coef_
}).sort_values('Coefficient', ascending=False)

rf_importance = pd.DataFrame({
    'Feature': X_train.columns,
    'Importance': rf_model.feature_importances_
}).sort_values('Importance', ascending=False)

xgb_importance = pd.DataFrame({
    'Feature': X_train.columns,
    'Importance': xgb_model.feature_importances_
}).sort_values('Importance', ascending=False)
```

```python
# Print the table of feature importance for each model
print('Linear Regression')
print(linear_importance)

print('Random Forest Regression')
print(rf_importance)

print('Support Vector Regression')
print('Feature importance is not available for this model')

print('XGBoost Regression')
print(xgb_importance)
```

```python
# Select the best model and make predictions on the test data
best_model = xgb_model
y_pred = best_model.predict(X_test)
```

```python
# Evaluate the performance of the best model
best_model_r2 = r2_score(y_test, y_pred)
best_model_mse = mean_squared_error(y_test, y_pred)
```

```python
print("Best Model R-squared:", best_model_r2)
print("Best Model MSE:", best_model_mse)
```

```python
# Get the feature importance for the best model
feature_importance = pd.DataFrame({'Feature': X_train.columns, 'Importance': best_model.feature_importances_})
feature_importance = feature_importance.sort_values(by=['Importance'], ascending=False)
```
